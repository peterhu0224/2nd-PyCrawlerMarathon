```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import pandas as pd
browser = webdriver.Chrome(executable_path=
r'C:\Users\hugohu\Dropbox\ShareWith_Hugo\CourseFile\chromedriver_win32\chromedriver')
browser.get("http://taqm.epa.gov.tw/taqm/tw/MonthlyAverage.aspx")
```

```python
selectSite = Select(browser.find_element_by_id("ctl05_ddlSite"))
selectSite.select_by_value('58')
selectYear = Select(browser.find_element_by_id("ctl05_ddlYear"))
selectYear.select_by_value('2018')
browser.find_element_by_id('ctl05_btnQuery').click()
```

```python
html_source = browser.page_source
soup = BeautifulSoup(html_source, 'html.parser')
table = soup.find('table', class_='TABLE_G')
browser.quit()
```

```python
dataList = []
c = 0
tList = []
for i in table.find('tbody').find_all('td'):
    tList.append(i.text.replace('\n','').replace('\xa0',''))
    c += 1
    while c==38:
        dataList.append(tList)
        tList = []
        c = 0
```

```python
result = {}
for i in dataList:
    title = '{}({})'.format(i[0], i[1])
    c = 0
    tList = []
    rList = {}
    for data in i[2:]:
        tList.append(data)
        c += 1
        while c%3 == 0:
            rList[tList[0]] = tList[1]
            c = 0
            tList = []
            break
    result[title] = rList
result
```
