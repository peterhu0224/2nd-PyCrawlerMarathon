```python
import requests
from bs4 import BeautifulSoup
url01 = 'https://www.dcard.tw/f'
url02 = 'https://www.zhihu.com/explore'
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.132 Safari/537.36'}
r1 = requests.get(url01)
r2 = requests.get(url02, headers=headers)
response1 = r1.text
response2 = r2.text
```

```python
soup1 = BeautifulSoup(response1, "html5lib")
print(soup1)
print(type(soup1))
```

```python
soup2 = BeautifulSoup(response2, "html5lib")
print(soup2)
print(type(soup2))
```
