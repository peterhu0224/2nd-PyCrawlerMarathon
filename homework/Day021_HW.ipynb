```python
import requests
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import pandas as pd
import time

browser = webdriver.Chrome(executable_path=
r'C:\Users\hugohu\Dropbox\ShareWith_Hugo\CourseFile\chromedriver_win32\chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
#1. 取出今天所有新聞
#爬蟲設定
now = time.localtime()[0:3]
selectSite = Select(browser.find_element_by_id("selY"))
selectSite.select_by_value(str(now[0]))
selectYear = Select(browser.find_element_by_id("selM"))
selectYear.select_by_value(str(now[1]))
selectYear = Select(browser.find_element_by_id("selD"))
selectYear.select_by_value(str(now[2]))
browser.find_element_by_id('button').click()
# 每秒自動往下滑，執行50次
for i in range(50):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight);")
#取網頁資料
soup = BeautifulSoup(browser.page_source, "html5lib")
df_today = pd.DataFrame()
#遍歷網頁資料
for block in soup.select('.part_list_2 h3'):
    date = (block.find('span').text)
    tag = block.find('em').text
    title = block.find('a').text
    #寫入df
    df_today = pd.concat([df_today, pd.Series([date, tag, title]).reset_index().T.iloc[1:]])
df_today.columns = ['date', 'em', 'title']
df_today['day'] = pd.to_datetime(df_today['date']).dt.day
df_today['hour'] = pd.to_datetime(df_today['date']).dt.hour
df_today[df_today['day']==now[2]]
```

```python
#2. 取出現在時間兩小時內的新聞
ts = list(time.localtime()[0:4])
print('now:',ts)
ts[3] = ts[3] - 2
print('within 2 hours:',ts)
df_today[(df_today['day']==ts[2])&(df_today['hour']>=ts[3])]
```

```python
#3. 取出三天前下午三點到五點的新聞
#爬蟲設定
ts = list(time.localtime()[0:4])
print('now:',ts)
daysago = 3
ts[2] = ts[2] - daysago
print('3daysago:',ts)
selectSite = Select(browser.find_element_by_id("selY"))
selectSite.select_by_value(str(ts[0]))
selectYear = Select(browser.find_element_by_id("selM"))
selectYear.select_by_value(str(ts[1]))
selectYear = Select(browser.find_element_by_id("selD"))
selectYear.select_by_value(str(ts[2]))
browser.find_element_by_id('button').click()
# 每秒自動往下滑，執行50次
for i in range(50):
    time.sleep(1)
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight);")
#取網頁資料
soup = BeautifulSoup(browser.page_source, "html5lib")
df_3daysago = pd.DataFrame()
#遍歷網頁資料
for block in soup.select('.part_list_2 h3'):
    date = (block.find('span').text)
    tag = block.find('em').text
    title = block.find('a').text
    #寫入df
    df_3daysago = pd.concat([df_3daysago, pd.Series([date, tag, title]).reset_index().T.iloc[1:]])
df_3daysago.columns = ['date', 'em', 'title']
df_3daysago['day'] = pd.to_datetime(df_3daysago['date']).dt.day
df_3daysago['hour'] = pd.to_datetime(df_3daysago['date']).dt.hour
df_3daysago[(df_3daysago['day']==ts[2])&(df_3daysago['hour']>=15)&(df_3daysago['hour']<=17)]
```
