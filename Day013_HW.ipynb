```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def testClawer(url):
    cookies18 = {'over18': '1'}
    r = requests.get(url, cookies=cookies18)
    soup = BeautifulSoup(r.text, "html5lib")
    urlHead = 'https://www.ptt.cc/'
    df = pd.DataFrame()
    for arts in soup.find_all(class_="title"):
        try:
            getUrl = urlHead+arts.find('a')['href']
            soup2 = BeautifulSoup(requests.get(getUrl, cookies=cookies18).text, "html5lib")
            tList = []
            for art in soup2.find_all(class_='article-meta-value'):
                tList.append(art.text)
            df = pd.concat([df, pd.Series(tList).reset_index().T[1:]])
        except:
            continue
    df.columns = ['Account', 'Kind', 'Title', 'Time']
    return df.dropna()
url = 'https://www.ptt.cc/bbs/NBA/index.html'
testClawer(url)
```

```python
url2 = 'https://www.ptt.cc/bbs/Gossiping/index.html'
testClawer(url2)
```
